```python
# Load the image
image_path = r"D:\\VIT Material\\VIT material\\Hepatoma Research Project\\Graph-Attention-VAE-cancer-detection-for-Hepatoma-Images-\Segmented-Images\Samp2\Samp2_image_18.png"
image = Image.open(image_path)
x = 255

# Separate the original and resized images from the array
original_image = np.array(image)
original_image = original_image / original_image.max()
resized_image = np.array(image.resize((128, 128)))

# Find coordinates of pixels with value 255 in the resized image
coordinates = np.column_stack(np.where(resized_image == x))

# Create a new image to highlight these coordinates
highlighted_image = np.zeros_like(resized_image)
highlighted_image[resized_image == x] = 255

# Convert coordinates list to numpy array
coordinates_array = np.array(coordinates)

# Display the original, resized, and highlighted images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(original_image, cmap='gray')
ax[0].set_title("Original")
ax[1].imshow(resized_image, cmap='gray')
ax[1].set_title("Resized to (128, 128)")
ax[2].imshow(highlighted_image, cmap='gray')
ax[2].set_title("Highlighted Centers")

plt.show()

# Print the coordinates array
print(coordinates_array)
```
